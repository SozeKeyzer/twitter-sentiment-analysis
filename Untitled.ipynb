```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re
import string
import nltk
import seaborn as sns
%matplotlib inline
```

```python
cols=["target","ids","date","flag","user","text"]
```

```python
df=pd.read_csv("training.1600000.processed.noemoticon.csv",encoding='ISO-8859-1')
df.columns=cols
```

```python
df.head()
```

```python
df=df.drop(columns=["ids","date","flag","user"])
df.head()
```

```python
df.info()
```

## Preprocessing dataset

```python
def remove_pattern(input_txt,pattern):
    r=re.findall(pattern,input_txt)
    for word in r:
        input_txt=re.sub(word,"",input_txt)
    return input_txt   
```

```python
#remove twitter handles
remove=np.vectorize(remove_pattern) #np.vectorize take a function that operates on individual elements and returns a fun that can be applied to NumPy arrays element by element
df['clean_txt']=remove(df['text'],"@[\w]*")
```

```python
df.head()
```

```python
#remove special characters,numbers and punctuations
df['clean_txt']=df['clean_txt'].str.replace("[^a-zA-Z#]"," ")
df.head()
```

```python
# remove short words
df['clean_txt']=df['clean_txt'].apply(lambda x: " ".join([w for w in x.split() if len(w)>3]))
```

```python
#tokenization
tokenized_text=df['clean_txt'].apply(lambda x:x.split())
tokenized_text.head()
```

```python
#stemming 
from nltk.stem.porter import PorterStemmer
stemmer=PorterStemmer()
tokenized_text=tokenized_text.apply(lambda sentence:[stemmer.stem(word) for word in sentence])
```

```python
tokenized_text.head()
```

```python
tokenized_text.info()
```

```python
#combine words into single sentence
for i in range(len(tokenized_text)):
    tokenized_text[i]=" ".join(tokenized_text[i])

df['clean_txt']=tokenized_text
df.head()
```

```python
# extract the hashtag
def hashtag_extract(tweets):
    hashtags=[]
    for tweet in tweets:
        ht=re.findall(r"#(\w+)",tweet)
        hashtags.append(ht)
    return hashtags    
```

```python
# extract hashtags from positive tweets
ht_positive =hashtag_extract(df['clean_txt'][df['target']==4])

# extract hashtags from negative tweets
ht_negative =hashtag_extract(df['clean_txt'][df['target']==0])
```

```python
ht_positive[0:6]
```

```python
# combine list
ht_positive =sum(ht_positive,[])
ht_negative=sum(ht_negative,[])
```

```python
ht_positive[0:6]
```

```python
freq=nltk.FreqDist(ht_positive)
# converting this dict. to dataframe
d=pd.DataFrame({'Hashtag':list(freq.keys()),
                'Count':list(freq.values())})
d.head()
```

```python
# select top 10 hashtags
d=d.nlargest(columns='Count',n=10)
plt.figure(figsize=(15,9))
sns.barplot(data=d,x='Hashtag',y='Count')
plt.show()
```

```python
freq=nltk.FreqDist(ht_negative)
# converting this dict. to dataframe
d=pd.DataFrame({'Hashtag':list(freq.keys()),
                'Count':list(freq.values())})
d.head()
```

```python
# select top 10 hashtags
d=d.nlargest(columns='Count',n=10)
plt.figure(figsize=(15,9))
sns.barplot(data=d,x='Hashtag',y='Count')
plt.show()
```

## Input Split

```python
# feature extraction
from sklearn.feature_extraction.text import CountVectorizer
bow_vectorizer=CountVectorizer(max_df=0.90,min_df=2,max_features=1000,stop_words='english')
bow=bow_vectorizer.fit_transform(df['clean_txt'])
```

```python
# split the imput for training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(bow,df['target'],random_state=42,test_size=0.25)
```

## Logistic Regg

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
#training
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
#testing
pred=model.predict(x_test)
print(classification_report(y_test, pred))
```
